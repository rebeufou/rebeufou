```python
btc['date'] = pd.to_datetime(btc['date'])
btc['date'] = pd.to_datetime(btc['date'], format= '%Y-%m-%d %I-%p')
```

```python
btc['date'].resample('D')
```

```python
btc['day'] = btc['date'].dt.day_name()
btc['numerical day'] = btc['date'].dt.day
btc = btc.sort_values(by='date', ascending = True)
btc
```

```python
btc_resample = btc.resample('D', on='date').mean()
```

```python
btc_resample
```

```python
btc_resample.plot()
```
