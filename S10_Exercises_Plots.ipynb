```python
#S10_4 Isotherm of ideal gas
while True:
  try:
    n = int(input("What is the number of points : "))
    if n>=1:
      break
  except:
    print("Error.Please enter a valide number :")
while True:
  try:
    T = int(input("What is the temperature : "))
    if T>=1:
      break
  except:
    print("Error.Please enter a valide number :")

x = np.linspace(2,10,n)
y = (0.08206*T)/x
plt.xlabel("Vm")
plt.ylabel("p")
plt.plot(x,y,'g',label="p = RT/Vm")
plt.title("Isotherm (ideal gas)")
plt.legend()
plt.show()
#plt.savefig("isotherm.png")
```

```python
#S10_5
import numpy as np
import matplotlib.pyplot as plt

n = int(input("What is the number of points: "))
z = int(input("What is the number of isotherms: "))
temperature = []

for i in range(z):
    temp = int(input("What is the value of temperature? "))
    temperature.append(temp)

x = np.linspace(2, 10, n)
plt.xlabel("Vm")
plt.ylabel("p")
plt.title("Isotherm (ideal gas)")

for i in range(z):
    y = (0.08206 * temperature[i]) / x
    plt.plot(x, y, label=("T = " + str(temperature[i]) + " K"))

plt.legend()
plt.show()
# plt.savefig("isotherm.png")
```

```python
#S10_6
import numpy as np
import matplotlib.pyplot as plt

n = int(input("What is the number of points: "))
a = float(input("What is the initial value : "))
b = float(input("What is the finale value : "))
x0 = int(input("What is the value of x0 : "))
s = float(input("What is the value of S : "))


x = np.linspace(a, b, n)
plt.title("Gaussian function")
plt.xlabel("x")
plt.ylabel("f(x)")
y = np.exp((-1/2)*((x-x0)**2)*(1/(s**2)))/np.sqrt(2*np.pi)
plt.plot(x, y, 'g')

plt.legend()
plt.show()
# plt.savefig("isotherm.png")
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = int(input("What is the number of points: "))
z = int(input("What is the number gaussians : "))
a = float(input("What is the initial value : "))
b = float(input("What is the finale value : "))
gaussians = []

for i in range(z):
    x0 = int(input("What is the value of x0 : "))
    s = float(input("What is the value of S : "))
    gaussians.append(x0)
    gaussians.append(s)

x = np.linspace(a, b, n)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gaussian function")

for i in range(0,len(gaussians)-2,2):
  y = np.exp((-1/2)*((x-gaussians[i])**2)*(1/(gaussians[i+1]**2)))/np.sqrt(2*np.pi)
  plt.plot(x, y, label=("x0 = " + str(gaussians[i]) + " , s = " + str(gaussians[i+1])))

plt.legend()
plt.show()
# plt.savefig("isotherm.png")
```
